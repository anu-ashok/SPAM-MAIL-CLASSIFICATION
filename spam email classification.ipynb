```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics  import accuracy_score
```

```python
df=pd.read_csv('spam.csv',encoding='latin-1')
df.head()
```

```python
data = df.fillna('')

data.head(10)
```

```python
data.info()
```

```python
data.shape
```

```python
data.loc[data['v1']=='spam','v1',]=0
data.loc[data['v1']=='ham','v1',]=1
```

```python
x=data['v2']
y=data['v1']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
print(y.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english', lowercase=True)
x_train_features=feature_extraction.fit_transform(x_train)
x_test_features=feature_extraction.transform(x_test)

y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
print(x_train)
```

```python
print(x_train_features)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train_features,y_train)
```

```python
prediction=model.predict(x_train_features)
accuracy_on_data=accuracy_score(y_train,prediction)
print('accuracy_on_training_data_is:',accuracy_on_data)
prediction=model.predict(x_test_features)
accuracy_test=accuracy_score(y_test,prediction)
print('accuracy_on_testing_data',accuracy_test)
```
